## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here

# Problem 1: How to aggregate spendings on customer level

pivot = orders.pivot_table(index='CustomerID',values='amount_spent', aggfunc=('sum')).reset_index()

pivot

# By using the pivot_table we're able to sum up the amount spent per unique customer id

# Problem 2: Check in which decile rank a customer is, by adding a new column with qcut that gives back 20 equally sized groups (deciles)

pivot['quantiles'] = pd.qcut(pivot['amount_spent'],q=20, labels=False)

display(pivot)

# Problem 3: Label Customers as VIP and Preferred Customers. Since we clustered Customers into 5% quantiles, 
# we can use those values in column quantile, to create a labelling column, that sets the label for the customer,
# based on the quantile he is in.

def customer_type(row):
    if (row['quantiles']==19):
        return 'VIP'
    elif(15 <= row['quantiles'] < 19):
        return 'Preferred Customer' 
    else:
        return 'Normal'

pivot['Customer_Type'] = pivot.apply(customer_type, axis=1)

display(pivot)
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

# Problem 1: How to aggregate spendings on country and customer level

countries = orders.pivot_table(index=['Country','CustomerID'],values='amount_spent', aggfunc=('sum')).reset_index()

countries

# Problem 2: Check in which decile rank a customer is, by adding a new column with qcut that gives back 20 equally sized groups (deciles)

countries['quantiles'] = pd.qcut(countries['amount_spent'],q=20, labels=False)

display(countries)

# Problem 3: Label Customers as VIP and Preferred Customers. Since we clustered Customers into 5% quantiles, 
# we can use those values in column quantile, to create a labelling column, that sets the label for the customer,
# based on the quantile he is in.

def customer_type(row):
    if (row['quantiles']==19):
        return 'VIP'
    elif(15 <= row['quantiles'] < 19):
        return 'Preferred Customer' 
    else:
        return 'Normal'

countries['Customer_Type'] = countries.apply(customer_type, axis=1)

display(countries)

#Problem 4: Count the number of VIP customers

vip = countries[countries['Customer_Type']=='VIP']

vip.pivot_table(index=['Country'], values=['CustomerID'], aggfunc=('count')).reset_index()


# Most VIP Customers come from the United Kingdom
```
